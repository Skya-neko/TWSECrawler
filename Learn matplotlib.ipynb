```python
#https://www.youtube.com/watch?v=46EsDY8V6lQ&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=5
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure() 
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')

plt.xlim((-1,2)) #限定x軸的範圍
plt.ylim((-2,3))
plt.xlabel('I am x') #x軸的名稱
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks) #換x軸的刻度


plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          ['really bad','bad','normal','good','really good'])

'''
(1)
plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])
          #加上r'$bad$'可以使出現的字變斜體，r'$really\ bad$'中間要加"\"編譯器才能辨識它是空格
(2)
plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          ['really bad',r'$bad\ \alpha$','normal','good','really good'])
          #在r'$[]$'的[]中打上\alpha，會被解讀成數學符號

'''


plt.show()
```

# [莫煩 6](https://www.youtube.com/watch?v=w83mFG5tyW4&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=6)


新增
* 指定脊 plt.gca().spines['']
* 設置脊顏色 plt.gca().spines[''].set_color('')
* 將刻度放置在自己指定的脊上 plt.gca().xaxis.set_ticks_position('')
* 設置脊於相對軸上的位置 plt.gca().spines[''].set_position('data',)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure()  #為了方便說明，叫這個figure為"圖1"
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')

plt.xlim((-1,2)) 
plt.ylim((-2,3))
plt.xlabel('I am x') 
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks) 

plt.yticks([-2,-1.8,-1,1.22,3], 
          [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])

#gca = 'get current axis'
ax = plt.gca() #圖1的ax
ax.spines['right'].set_color('none') #spines代表圖形上下左右4邊。此句表示將右脊刪掉 (spine為脊椎，在這自定義為右脊)
ax.spines['top'].set_color('none') #刪掉上脊
ax.xaxis.set_ticks_position('bottom') #將x軸刻度固定在下脊
ax.yaxis.set_ticks_position('left') #將y軸刻度固定在左脊

ax.spines['bottom'].set_position(('data',0)) #將下脊固定在('data',0)的位置，data表示圖形預設的xy軸，現在是橫軸(下脊)
                                             #，故('data',0)為相對於橫軸 的縱軸(y軸) 上 0 的位置 
    
ax.spines['left'].set_position(('data',0)) #將左脊固定在('data',0)的位置，data表示圖形預設的xy軸，現在是縱軸(左脊)
                                             #，故('data',0)為相對於縱軸 的橫軸(x軸) 上 0 的位置 

plt.show()
```

# [莫煩 8](https://www.youtube.com/watch?v=p7FoQqWk_Uo&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=8)

新增
* 圖上註解(annotation)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5))
plt.plot(x, y)
#plt.scatter(x,y)--->會變成點陣圖

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))


x0 = 1
y0 = 2*x0+1
plt.scatter(x0,y0,s=50,color='r') #s-->size 
plt.plot([x0,x0],[y0,0],'k--',lw=2.5) #[起點],[終點]--->連成一條線，'k--'-->k表示black，--表示線的樣式，lw-->線的寬度

plt.annotate(r'$2x+1=%s$' % y0,xy=(1.2,y0),xycoords='data',fontsize=16,arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))


plt.text(-3.7,3,r'$This\ is\  text. $',fontdict={'size':16,'color':'r'})

plt.show()
```

# [莫煩_15 subplot](https://www.youtube.com/watch?v=Zm1cDw7DnUA&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=15)

新增
* 顯示多張圖於一個figure

```python
#1
import matplotlib.pyplot as plt

plt.figure()

plt.subplot(2,2,1) #(2欄,2列,第1個位置)
plt.plot([0,1],[0,1]) 
'''
前[]是x的list，後[]是y的list，表示0~1，
也就是當x=0,y=0
x=0.1,y=0.1，由無數個點從(0,0)到(1,1)
'''
plt.subplot(2,1,1) #(2欄,2列,第2個位置)
plt.plot([0,1],[0,2])


plt.subplot(2,3,4) #(2欄,2列,第3個位置)
plt.plot([0,1],[0,3])


plt.subplot(2,3,5) #(2欄,2列,第4個位置)
plt.plot([0,1],[0,4])

plt.show()
```

```python
#2
import matplotlib.pyplot as plt

plt.figure()

plt.subplot(2,2,1) #(2欄,2列,第1個位置)
plt.plot([0,1],[0,1]) 
'''

前[]是x的list，後[]是y的list，表示0~1，
也就是當x=0,y=0
x=0.1,y=0.1，由無數個點從(0,0)到(1,1)
'''
plt.subplot(2,1 ,2) #(2欄,2列,第2個位置)
plt.plot([0,1],[0,2])


plt.subplot(2,2,3) #(2欄,2列,第3個位置)
plt.plot([0,1],[0,3])


plt.subplot(2,2,4) #(2欄,2列,第4個位置)
plt.plot([0,1],[0,4])

plt.show()
```
