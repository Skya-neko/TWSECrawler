基礎

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure() #figure~下一個figure前 的設定  都是屬於一個figure的
plt.plot(x,y1)


'''
plt.figure(num=None,figsize=(n1,n2))
num:图像编号或名称，数字为编号 ，字符串为名称
figsize:指定figure的宽和高，单位为英寸；
'''
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=10.0,linestyle='--')


plt.show()
```

# [莫煩 5](https://www.youtube.com/watch?v=46EsDY8V6lQ&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=5)

新增
* 軸範圍(xlim)
* 軸名稱(xlabel)
* 軸刻度(xticks)

```python
#https://www.youtube.com/watch?v=46EsDY8V6lQ&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=5
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure() 
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')

plt.xlim((-1,2)) #限定x軸的範圍
plt.ylim((-2,3))
plt.xlabel('I am x') #x軸的名稱
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks) #換x軸的刻度


plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          ['really bad','bad','normal','good','really good'])

'''
(1)
plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])
          #加上r'$bad$'可以使出現的字變斜體，r'$really\ bad$'中間要加"\"編譯器才能辨識它是空格
(2)
plt.yticks([-2,-1.8,-1,1.22,3], #換y軸的刻度及名稱，上下兩相對應
          ['really bad',r'$bad\ \alpha$','normal','good','really good'])
          #在r'$[]$'的[]中打上\alpha，會被解讀成數學符號

'''


plt.show()
```

# [莫煩 6](https://www.youtube.com/watch?v=w83mFG5tyW4&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=6)


新增
* 指定脊 plt.gca().spines['']
* 設置脊顏色 plt.gca().spines[''].set_color('')
* 將刻度放置在自己指定的脊上 plt.gca().xaxis.set_ticks_position('')
* 設置脊於相對軸上的位置 plt.gca().spines[''].set_position('data',)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure()  #為了方便說明，叫這個figure為"圖1"
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')

plt.xlim((-1,2)) 
plt.ylim((-2,3))
plt.xlabel('I am x') 
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks) 

plt.yticks([-2,-1.8,-1,1.22,3], 
          [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])

#gca = 'get current axis'
ax = plt.gca() #圖1的ax
ax.spines['right'].set_color('none') #spines代表圖形上下左右4邊。此句表示將右脊刪掉 (spine為脊椎，在這自定義為右脊)
ax.spines['top'].set_color('none') #刪掉上脊
ax.xaxis.set_ticks_position('bottom') #將x軸刻度固定在下脊
ax.yaxis.set_ticks_position('left') #將y軸刻度固定在左脊

ax.spines['bottom'].set_position(('data',0)) #將下脊固定在('data',0)的位置，data表示圖形預設的xy軸，現在是橫軸(下脊)
                                             #，故('data',0)為相對於橫軸 的縱軸(y軸) 上 0 的位置 
    
ax.spines['left'].set_position(('data',0)) #將左脊固定在('data',0)的位置，data表示圖形預設的xy軸，現在是縱軸(左脊)
                                             #，故('data',0)為相對於縱軸 的橫軸(x軸) 上 0 的位置 

plt.show()
```

# [莫煩 7](https://www.youtube.com/watch?v=dGZyoX72iEg&list=PLXO45tsB95cKiBRXYqNNCw8AUo6tYen3l&index=7)

新增
* 圖例(legend)


```python
#1
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure()  #為了方便說明，叫這個figure為"圖1"


plt.xlim((-1,2)) 
plt.ylim((-2,3))
plt.xlabel('I am x') 
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks) 

plt.yticks([-2,-1.8,-1,1.22,3], 
          [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])

plt.plot(x,y2,label='up') #label是給線取一個名字
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--',label='down')
plt.legend(loc='center right') 

'''
loc-->location 可使用的參數有
單獨
"best"(最好位置)
"cneter"

搭配
"upper"  "right"
"lower"  "left"
"center" "center"

'''

l1, = plt.plot(x,y2,label='up') #label是給線取一個名字
l2, = plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--',label='down')
plt.legend(handles=[l1,l2,],loc='center right') 

#handles=,labels=,
plt.show()
```
